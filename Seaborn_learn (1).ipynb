```python
sns.get_dataset_names()
```

```python
mpg = sns.load_dataset('mpg')
```

```python
sns.set_theme()
```

```python
sns.set_style('dark') #white, whitegrid, darkgrid
```

```python
sns.set_context('paper', font_scale = 1.1)  #paper,talk,poster
```

```python
mpg.corr(numeric_only=True)
```

```python
sns.heatmap(mpg.corr(numeric_only=True))
```

```python
#plt.figure(figsize = (15,5))
sns.scatterplot(data = mpg, x = 'weight', y = 'mpg', hue = 'origin', palette = 'mako')
sns.despine(left = True, right = False)
```
